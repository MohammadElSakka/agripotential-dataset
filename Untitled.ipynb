```python
import os
import numpy as np
from tqdm import tqdm

from rasterio import rasterio
from rasterio.windows import Window

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, BoundaryNorm

from scripts.dataset import Dataset
```

```python

def brighten(data: np.ndarray) -> np.ndarray:
    alpha=0.08
    beta=0
    return np.clip(alpha*data+beta, 0, 255)
    
def normalize(arr):
    arr_min, arr_max = (arr.min(), arr.max())
    return ((arr-arr_min)/((arr_max - arr_min)))
```

```python
data_path = {}

sentinel2_paths = ["data/raw_data/"+f for f in os.listdir("data/raw_data") if f.endswith("tif")]

sentinel2_paths
```

```python
for path in sentinel2_paths:
    data_path[int(path.split("_")[-2])] = path
data_path
```

```python
dataset = Dataset(download=False, ind_conf=3, iddiz=2.5 , icucs=3)
train_mask, train_positions, val_mask, val_positions, vit_potential, binary_mask = dataset.process()
```

```python
grid_mask = np.zeros((train_mask.shape[0], train_mask.shape[1], 3), dtype=np.uint8)
grid_mask[train_mask == 1] = [0, 0, 255]
grid_mask[val_mask == 1] = [255, 0, 0]    
plt.imshow(grid_mask)
plt.show()
```

```python
potential = np.argmax(vit_potential, axis=-1)
potential = np.where(binary_mask, potential, np.nan)
colors = ["red", "orange", "yellow", "lightgreen", "green"]
cmap = ListedColormap(colors)
bounds = np.linspace(0, 5, 6)
norm = BoundaryNorm(bounds, cmap.N)
plt.imshow(potential, cmap=cmap, norm=norm, interpolation="none")
plt.show()
```

```python
color_map = {
    1: [1.0, 0.0, 0.0],          # Red
    2: [1.0, 165/255, 0.0],      # Orange
    3: [1.0, 1.0, 0.0],          # Yellow
    4: [144/255, 238/255, 144/255],  # Light Green
    5: [0.0, 128/255, 0.0]       # Green
}
```

```python
train_data = []
val_data = []
block_size = 256

idx = np.random.randint(0, 1200)

for month in tqdm(range(1, 13)):
    for i, j in train_positions[idx:]:
        patch = rasterio.open(data_path[month]).read(window=Window(j, i, block_size, block_size))
        b = normalize(brighten(patch[0]))
        g = normalize(brighten(patch[1]))
        r = normalize(brighten(patch[2]))
        nir = normalize(brighten(patch[6]))
        rgb = np.dstack([r, g, b])
        false_rgb = np.dstack([nir, r, g])

        potential = np.argmax(vit_potential, axis=-1)
        potential = potential[i:i+block_size, j:j+block_size]
        potential_rgb = np.zeros((256,256,3))
                                 
        for value, color in color_map.items():
            potential_rgb[potential == value] = color
            
        plt.imshow(rgb)
        plt.show()
        plt.imshow(potential_rgb)
        plt.show()
        break
    break
    
#     train_data.append(train_mask[i:i+block_size, j:j+block_size])

```
